```python
import pandas as pd
```

```python
#Importando dados
file_path = '/content/vendas_de_carros_com_data.csv'
sales_data = pd.read_csv(file_path)
```

```python
#Exibir as primeiras linhas do conjunto de dados
sales_data.head()
```

```python
# Imputar valores ausentes com mediana
sales_data['preco'].fillna(sales_data['preco'].median(), inplace=True)
sales_data['potencia'].fillna(sales_data['potencia'].median(), inplace=True)
```

Agora que tratamos os valores ausentes e garantimos que as colunas estão com os tipos de dados corretos, podemos prosseguir com a Análise Exploratória de Dados (AED). Isso envolve criar visualizações e calcular estatísticas descritivas para entender melhor as características do conjunto de dados.

```python
import numpy as np
#Calcule estatísticas descritivas para colunas numéricas
descriptive_stats = sales_data.describe(include=[np.number])
```

```python
print(descriptive_stats)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the matplotlib figure
plt.figure(figsize=(12, 8))

# Draw a histogram of 'preco'
sns.histplot(sales_data['preco'], bins=30, kde=True)

# Show the plot
plt.show()
```

```python
# Calcule estatísticas descritivas para colunas numéricas
descriptive_stats = sales_data.describe(include=[np.number])
descriptive_stats
```

```python
# Set up the matplotlib figure
plt.figure(figsize=(12, 8))

# Draw a histogram of 'km'
sns.histplot(sales_data['km'], bins=30, kde=True)

# Show the plot
plt.title('Histograma de Quilometragem dos Carros')
plt.xlabel('Quilometragem (km)')
plt.ylabel('Frequência')
plt.xticks([0, 60000, 120000, 180000, 240000, 300000, 360000, 420000, 480000, 540000])
plt.show()
```

```python
# Set up the matplotlib figure
plt.figure(figsize=(12, 8))

# Draw a histogram of 'potencia'
sns.histplot(sales_data['potencia'], bins=30, kde=True)

# Show the plot
plt.title('Histograma de Potência dos Carros')
plt.xlabel('Potência (L)')
plt.ylabel('Frequência')
plt.show()
```

```python
# Set up the matplotlib figure
plt.figure(figsize=(12, 8))

# Draw a histogram of 'ano'
sns.histplot(sales_data['ano'], bins=30, kde=True)

# Show the plot
plt.title('Histograma de Ano de Fabricação dos Carros')
plt.xlabel('Ano de Fabricação')
plt.ylabel('Frequência')
plt.show()
```

```python
# Set up the matplotlib figure
plt.figure(figsize=(12, 8))

# Draw a scatter plot of 'preco' vs 'km'
sns.scatterplot(x='km', y='preco', data=sales_data)

# Show the plot
plt.title('Gráfico de Dispersão entre Preço e Quilometragem dos Carros')
plt.xlabel('Quilometragem (km)')
plt.ylabel('Preço (R$)')
plt.show()
```

```python
# Set up the matplotlib figure
plt.figure(figsize=(14, 8))

# Draw a boxplot of 'preco' by 'marca'
sns.boxplot(x='marca', y='preco', data=sales_data)

# Rotate x labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.title('Boxplot de Preço por Marca')
plt.xlabel('Marca')
plt.ylabel('Preço (R$)')
plt.show()
```

Correlação entre variáveis numéricas

Mostra quanto uma variável é implicada ou afetada por outra variáve

```python
# Calculate the correlation matrix
correlation_matrix = sales_data.corr()

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Draw a heatmap of the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")

# Show the plot
plt.title('Matriz de Correlação entre Variáveis Numéricas')
plt.show()
```

A matriz de correlação mostra como as variáveis numéricas do conjunto de dados estão relacionadas entre si. Os valores de correlação variam de -1 a 1, onde:

1 indica uma correlação positiva perfeita, -1 indica uma correlação negativa perfeita, 0 indica nenhuma correlação. Neste caso, observamos que:

Existe uma correlação negativa moderada entre preco e km de aproximadamente -0.39, sugerindo que carros com maior quilometragem tendem a ter preços mais baixos. Existe uma correlação positiva fraca entre preco e ano de aproximadamente 0.26, sugerindo que carros mais novos tendem a ter preços mais altos. Existe uma correlação negativa fraca entre km e ano de aproximadamente -0.32, sugerindo que carros mais novos tendem a ter menos quilometragem.


```python
# Set up the matplotlib figure
plt.figure(figsize=(14, 8))

# Draw a bar plot of car count by 'marca'
sns.countplot(y='marca', data=sales_data, order=sales_data['marca'].value_counts().index)

# Show the plot
plt.title('Contagem de Carros por Marca')
plt.xlabel('Contagem')
plt.ylabel('Marca')
plt.show()
```

```python
# Set up the matplotlib figure
plt.figure(figsize=(14, 8))

# Draw a scatter plot of 'ano' vs 'preco' and color by 'marca'
sns.scatterplot(x='ano', y='preco', hue='marca', data=sales_data, palette='Set2')

# Show the plot
plt.title('Relação entre Ano de Fabricação e Preço para diferentes Marcas')
plt.xlabel('Ano de Fabricação')
plt.ylabel('Preço (R$)')
plt.show()
```

```python
# Set up the matplotlib figure
plt.figure(figsize=(14, 8))

# Draw a scatter plot of 'preco' vs 'potencia'
sns.scatterplot(x='potencia', y='preco', data=sales_data, alpha=0.6)

# Show the plot
plt.title('Relação entre Preço e Potência dos Carros')
plt.xlabel('Potência (L)')
plt.ylabel('Preço (R$)')
plt.show()
```

```python
# Group the data by 'ano' and calculate the average 'preco' for each year
average_price_by_year = sales_data.groupby('ano')['preco'].mean().reset_index()

# Set up the matplotlib figure
plt.figure(figsize=(14, 8))

# Draw a line plot of average 'preco' over 'ano'
sns.lineplot(x='ano', y='preco', data=average_price_by_year)

# Show the plot
plt.title('Tendência do Preço Médio dos Carros ao Longo dos Anos')
plt.xlabel('Ano de Fabricação')
plt.ylabel('Preço Médio (R$)')
plt.show()
```

```python
# Group the data by 'ano' and calculate the average 'preco' for each year
average_price_by_year = sales_data.groupby('data')['preco'].mean().reset_index()

# Set up the matplotlib figure
plt.figure(figsize=(14, 8))

# Draw a line plot of average 'preco' over 'ano'
sns.lineplot(x='data', y='preco', data=average_price_by_year)

# Show the plot
plt.title('Tendência do Preço Médio dos Carros ao Longo das Datas das Compras')
plt.xlabel('Ano de Fabricação')
plt.ylabel('Preço Médio (R$)')
plt.show()
```
